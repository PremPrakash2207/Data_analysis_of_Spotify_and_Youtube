```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
data = pd.read_csv('/content/Spotify Youtube Dataset.csv', engine='python', encoding='ISO-8859-1')
```

```python
display(data)
```

```python
data.columns
```

```python
data.info()
```

```python
data.drop(columns=['Url_spotify', 'Uri', 'Url_youtube'] , inplace=True)
```

```python
display(data)
```

```python
data.isnull().sum()
```

```python
#filling the missing values with 0 in likes and comments colum
data['Likes'] = data['Likes'].fillna(0)
data['Comments'] = data['Comments'].fillna(0)
```

```python
data.isnull().sum()
```

```python
data.dropna(inplace = True)
```

```python
data.isnull().sum()
```

```python
data.info()
```

```python
#Top 10 Artist with highest views Youtube
data.groupby('Artist')['Views'].sum().sort_values(ascending=False).head(10)
```

```python
#top 10 highest track with highest stream on Spotify
data.groupby('Track')['Stream'].sum().sort_values(ascending=False).head(10)
```

```python
data.groupby('Track')['Stream'].sum().sort_values(ascending=False).head(5)
```

```python
#what the most common album types on spotify? How many track belong to each album type?
data.Album_type.unique()
Album_type = data.Album_type.value_counts()
Album_type
```

```python
#draw a pie chart
plt.pie(Album_type, labels= Album_type.index, autopct='%1.1f%%', startangle=60, colors='grb', shadow=True, explode=(0.05,0.05,0.05), pctdistance=0.55)
plt.show()
```

```python
#how do the average views, likes, and comments are compared between different album types
df = data.groupby('Album_type')[['Views', 'Likes', 'Comments']].mean()
df
```

```python
df_reset = df.reset_index()
df.head()
```

```python
type(df_reset)
#melt unpivot a dataframe
df_melted = pd.melt(df_reset, id_vars ='Album_type', var_name='Attribute', value_name='Values')
df_melted
```

```python
#draw barplot
plt.figure(figsize=(9,4))
sns.barplot(x='Album_type', y='Values', hue='Attribute', data=df_melted)
```

```python
#top 5 youtube channels based on the views
C_views = data.groupby('Channel')['Views'].sum().sort_values(ascending=False).head(5)
C_views
```

```python
C_views_df = C_views.reset_index()
C_views_df.columns = ['Channel', 'Views']
sns.barplot(x= 'Views', y= 'Channel', data=C_views_df, color= 'black');
plt.xlabel('Views');
plt.ylabel('Channel');
plt.title('Top 5 Youtube channels based on the views');
```

```python
#the top most track based on views
data.sort_values(by='Views', ascending=False).head(1)
```

```python
#which top7 tracks has highest like to view ration on youtube?
track_lv = data[['Track', 'Views', 'Likes']]
track_lv['LV_ratio'] = track_lv['Likes']/track_lv['Views']
track_lv.head()
```

```python
track_lv.sort_values(by='LV_ratio', ascending=False).head(7)
```

```python
#which top7 tracks has lowest like to view ration on youtube?
track_lv.sort_values(by='LV_ratio', ascending=True).head(7)
```

```python
#top album having track with maximum dancebility
T_danceability = data.groupby('Album')['Danceability'].sum().sort_values(ascending=False)
T_danceability
data[data.Album=='Greatest Hits']
```
